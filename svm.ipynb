```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import os
x = []
y = []
for category in os.listdir('C:/Users/Shubham/Desktop/text_classification/20_newsgroups'):
    for document in os.listdir(f"C:/Users/Shubham/Desktop/text_classification/20_newsgroups/{category}"):
        with open(f"C:/Users/Shubham/Desktop/text_classification/20_newsgroups/{category}/{document}", 'r') as f:
            x.append(f.read())
            y.append(category)
```

```python
print(f"Length of Feature list = {len(x)}")
print(f"Length of Target list = {len(y)}")
```

```python
from re import sub
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
# Function to convert into simple pos tags
def get_simple_pos(tag):    
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return 'r'
    else:
        return wordnet.NOUN
```

```python
X = []
for i in range(len(x)):
    a = sub('[^a-zA-Z]', " ", x[i])
    a = a.strip().split()
    lemmatizer = WordNetLemmatizer()
    a = [lemmatizer.lemmatize(word, pos = get_simple_pos(pos_tag([word])[0][1])).lower() for word in a if not word.lower() in set(stopwords.words('english'))]
    a = " ".join(a)
    X.append(a)
    print(i, end=" ")
```

```python
from sklearn.preprocessing import LabelEncoder
y = np.array(y)
label_y = LabelEncoder()
y_final = label_y.fit_transform(y)
y_final.shape, label_y.inverse_transform([2])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

```python
score = []
for features in range(1000, 4001, 100):
    tfidf = TfidfVectorizer(max_features= features, ngram_range= (1,3))
    temp = tfidf.fit_transform(X).toarray()
    clf_temp = SVC(kernel= 'linear')
    score.append(cross_val_score(estimator=clf_temp, X=temp, y=y_final, cv=3).mean())
    print(features, end = " ")
    plt.figure(figsize=(10, 10))
    plt.scatter(range(1000, features +1, 100), score, color='blue')
    plt.plot(range(1000, features +1, 100), score, color='red')
    plt.xlabel('Number of Features')
    plt.ylabel('CV Accuracy Score')
    plt.grid()
    # plt.axis([1400, 3100, 0.89, 0.912])
    plt.title("SVM")
    plt.show()
```
